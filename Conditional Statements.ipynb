# Match-case statement

```python
number=2
match number:
    case 1:
        print('one')
    case 2|3:
        print('two or three')
    case _:
        print('other number')
```

# closest number using conditional statements

```python
n=13
m=4
q=n//m
low=m*q
high=m*(q+1)
if abs(n-low)<abs(n-high):
    print(low)
else:
    print(high)
```
