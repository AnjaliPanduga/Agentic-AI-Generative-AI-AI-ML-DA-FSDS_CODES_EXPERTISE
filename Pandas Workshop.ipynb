# univariate analysis (plot the graph using 1 variable)


```python
vis1=sns.distplot(stats['InternetUsers'])
plt.show()
```

```python
vis3=sns.distplot(stats['InternetUsers'],bins=10)
plt.show()
```

# Bivariate analysis (plot the graph using 2 variable)


```python
# Box Plot
vis4=sns.boxplot(data=stats,x='IncomeGroup',y='BirthRate')
plt.show()
```

```python
vis5=sns.lmplot(data=stats,x='InternetUsers',y='BirthRate')
plt.show()
```

```python
vis5=sns.lmplot(data=stats,x='InternetUsers',y='BirthRate',fit_reg=False)
plt.show()
```

```python
vis5=sns.lmplot(data=stats,x='InternetUsers',y='BirthRate',fit_reg=True,hue='IncomeGroup') #hue gives color information
plt.show()
```
